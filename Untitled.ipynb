```python
import pandas as pd
import numpy as np 
from  matplotlib import pyplot as plt
```

```python
#data  = pd.read_json('dpc-covid19-ita-province-latest.json')
data =  pd.read_json('dpc-covid19-ita-regioni.json')
```

```python
reg_l = data.denominazione_regione.unique()
```

```python
data_marche = data.loc[data['denominazione_regione']=="Marche"]
```

```python
data_marche.columns
```

```python
fig = plt.figure(figsize=(8,8))
fig.add_subplot(2,1,1)
plt.plot(data_marche.totale_casi.values)
fig.add_subplot(2,1,2)
plt.plot(np.log(data_marche.totale_casi.values))
```

```python
for r in reg_l:
    data_reg = data.loc[data['denominazione_regione']==r]
    fig = plt.figure(figsize=(8,8))
    fig.add_subplot(2,1,1)
    plt.title(r)
    plt.plot(data_reg.totale_casi.values)
    fig.add_subplot(2,1,2)
    plt.plot(np.log(data_reg.totale_casi.values))
    
```
